```python
#Plotting all the neighbourhood group
```

```python
sns.countplot(x = 'neighbourhood_group',data = Airbnb, palette="plasma")
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Neighbourhood Group')
```

```python
#Checking whether restaurant is delivering online or not by using Graph
```

```python
sns.countplot(x='room_type', data=Airbnb, palette="plasma" )
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Restaurants delivering online or Not')
```
